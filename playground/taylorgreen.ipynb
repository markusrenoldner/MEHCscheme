```python
M = 50
u = np.zeros([50,50])
pi = np.pi
v = 1
for t in range(0,2,M):
    t = 0
    for x in range(0,M):   
        for y in range(0,M):
            u[x,y] = -np.sin(pi*x)*np.cos(pi*x)*np.exp(-2*pi**2*v*t)

plt.plot(u)

print(u[0,:])
```
