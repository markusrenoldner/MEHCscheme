```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

- Re1600, E1, E2
- Re1600, Delta E1, Delta E2, dissipation of E1 and E2

```python
filename = 'periodic-cons-testmooslechner'
# filename = 'periodic-cons-test'

data   = np.genfromtxt("../out/rawdata/"+filename+".txt", delimiter=',').flatten()
time =    data[0::7]#[:50]
K1 =      data[1::7]
K2 =      data[2::7]
deltaK1 = data[3::7]#[:50]
diss1  =  data[4::7]#[:50]
deltaK2 = data[5::7]#[:50]
diss2  =  data[6::7]#[:50]
```

```python
plt.plot(time,K1,"m",label="$K_1$")
plt.plot(time,K2,"c",label="$K_2$")
plt.legend()
plt.grid()
plt.xlabel("$t$")
plt.ylabel("$K_i$")
plt.tight_layout()
plt.savefig("../out/plots/"+filename+"-energy.pdf")
```

```python
plt.plot(time,deltaK1,"k",label="$\Delta K_1$")
plt.plot(time,diss1  ,"m:",label="$2\mathcal{E}_{\zeta}$Re",linewidth=4.0)
plt.plot(time,deltaK2,"c",label="$\Delta K_2$")
plt.plot(time,diss2  ,"k:",label="$2\mathcal{E}_{\omega}$Re",linewidth=4.0)
plt.legend()
plt.grid()
plt.xlabel("$t$")
plt.ylabel("$\Delta K_i$, Dissipation")
plt.tight_layout()
plt.savefig("../out/plots/"+filename+"-energydissipation.pdf")
```
