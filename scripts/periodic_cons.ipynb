```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

### conservation: periodic domain, Re=$\infty$, $\Delta t=0.05$

```python
data = genfromtxt("../out/rawdata/periodic-cons-invisc.txt", delimiter=',').flatten()

time = data[0::7]
m1 =   data[1::7]
m2 =   data[2::7]
k1 =   data[3::7]
k2 =   data[4::7]
h1 =   data[5::7]
h2 =   data[6::7]
```

```python
plt.plot(time,m1,"c",label="div$(u)$")
plt.plot(time,m2,"m",label="div$(v)$")
plt.legend()
plt.grid()
plt.ylim(0,0.000000000001)
plt.xlabel("time")
plt.tight_layout()
plt.ylabel("div$(u)$, div$(v)$")
# plt.savefig("../out/plots/periodic-cons-div-invisc.pdf")
```

```python
plt.plot(time,k1,label="$K_1$")
plt.plot(time,k2,label="$K_2$")
plt.legend()
plt.grid()
plt.ylim(-0.00000000001,0.00000000001)
plt.xlabel("time")
plt.tight_layout()
plt.ylabel("$K_1$, $K_2$")
plt.savefig("../out/plots/periodic-cons-K-invisc.pdf")
```

```python
plt.plot(time,h1,label="$H_1$")
plt.plot(time,h2,label="$H_2$")
plt.legend()
plt.grid()
plt.tight_layout()
plt.ylim(-0.00000000001,0.00000000001)
plt.xlabel("time")
plt.ylabel("$H_1$, $H_2$")
plt.savefig("../out/plots/periodic-cons-H-invisc.pdf")
```

### conservation: periodic domain, Re=$100$, $\Delta t=0.05$

```python
data = genfromtxt("../out/rawdata/periodic_cons_Re100.txt", delimiter=',').flatten()

time = data[0::7]
m1 =   data[1::7]
m2 =   data[2::7]
k1 =   data[3::7]
k2 =   data[4::7]
h1 =   data[5::7]
h2 =   data[6::7]
```

```python
plt.plot(time,m1,label="div$(u)$")
plt.plot(time,m2,label="div$(v)$")
plt.legend()
plt.grid()
plt.ylim(0,0.000000000001)
plt.xlabel("time")
plt.ylabel("div$(u)$, div$(v)$")
plt.tight_layout()
plt.savefig("../out/plots/periodic-cons-div-Re100.pdf")
# plt.show()
```

```python
plt.plot(time,k1,label="$\Delta K_1-2E_2/$Re")
plt.plot(time,k2,label="$\Delta K_2-2E_1/$Re")
plt.legend()
plt.grid()
plt.ylim(-0.000000000001,0.000000000001)
# plt.ylim(-2,2)
plt.xlabel("time")
plt.tight_layout()
plt.ylabel("$K_1$, $K_2$")
plt.savefig("../out/plots/periodic-cons-K-Re100.pdf")
```

```python
plt.plot(time,h1,label="$\Delta H_1-D$")
plt.plot(time,h2,label="$\Delta H_2-D$")
plt.legend()
plt.grid()
plt.ylim(-0.00000000001,0.00000000001)
# plt.ylim(-2,2)
plt.xlabel("time")
plt.tight_layout()
plt.ylabel("$H_1$, $H_2$")
plt.savefig("../out/plots/periodic-cons-H-Re100.pdf")
```
