### wikipedia TGV solution with Re=\infty 

=> in this case u is time-independent and curl(f) = 0

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.arange(0, 2*np.pi, 0.2)
y = np.arange(0, 2*np.pi, 0.2)
X, Y = np.meshgrid(x, y)

ux,uy = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
 
color = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
M = np.sqrt(ux*ux+uy*uy) # magnitude
plt.quiver(X, Y, ux,uy,M,cmap=plt.cm.jet)
plt.xticks(np.arange(0, 2*np.pi+0.01, np.pi), ['0','π','2π'])
plt.yticks(np.arange(0, 2*np.pi+0.01, np.pi), ['0','π','2π'])
plt.colorbar(cmap=plt.cm.jet)
plt.show()
```

### rescaling for [-1,1]^3

```python
x = np.arange(-1,1, 0.2/np.pi)
y = np.arange(-1,1, 0.2/np.pi)
X, Y = np.meshgrid(x, y)

ux,uy = np.array([np.cos(X*np.pi)*np.sin(Y*np.pi),
                 -np.sin(X*np.pi)*np.cos(Y*np.pi)])
 
color = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
M = np.sqrt(ux*ux+uy*uy) # magnitude
plt.quiver(X, Y, ux,uy,M,cmap=plt.cm.jet)
plt.xticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.yticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.colorbar(cmap=plt.cm.jet)
plt.show()
```
