```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

```python
data   = np.genfromtxt("../out/rawdata/dirichlet-conv-invisc.txt", delimiter=',').flatten()

refstep = data[0::4][-5:]
u =       data[1::4][-5:]
v =       data[2::4][-5:]
uvdiff =  data[3::4][-5:]


plt.plot(refstep,u,"m.-",label="u, invisc")
plt.plot(refstep,v,"c.-",label="v, invisc")
# plt.plot(refstep,uvdiff,"k.-",label="u-v, Re=1")
plt.plot(refstep,[0.05*refstep[i]**1 for i in range(len(refstep))],"k-.",label="h^1")
plt.plot(refstep,[0.02*refstep[i]**2 for i in range(len(refstep))],"k--",label="h^2")
# plt.plot(refstep,[0.01*refstep[i]**3 for i in range(len(refstep))],"k-",label="h^3")

plt.legend()
plt.title("Dirichlet")
plt.yscale("log")
plt.xscale("log")
plt.ylabel("L2error")
# plt.xlabel("h")
plt.tight_layout()
plt.grid()
# plt.savefig("../out/plots/dirichlet-conv-invisc.pdf")
```
