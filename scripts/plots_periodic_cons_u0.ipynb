```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
z = np.arange(-1, 1, 0.1)
X,  Z = np.meshgrid(x,  z)

pi = np.pi

ux,uy,uz = np.array([np.cos(2*Z*pi),np.sin(2*pi*Z),np.sin(2*pi*X)])
 
# color = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
M = np.sqrt(ux*ux+uz*uz) # magnitude
plt.quiver(X, Z, ux,uz,M,cmap=plt.cm.jet)
plt.xticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.yticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.colorbar(cmap=plt.cm.jet)
plt.tight_layout()
# plt.show()
plt.savefig("../out/plots/periodic-cons-u0-xz.pdf")
```
