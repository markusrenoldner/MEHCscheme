```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

### dirichlet: ess BC on dual space 

```python
data = genfromtxt("../out/dirichlet_dual_conv_inviscid.txt", delimiter=' ').flatten()
v =        data[1::5]
u =        data[2::5]
uvdiff =   data[3::5]
time =     data[4::5]
meshsize = data[0::5]
```

```python
plt.plot(meshsize,uvdiff,"c.-",label="u-v")
plt.plot(meshsize,u,"m.-"     ,label="u")
plt.plot(meshsize,v,"k.--"    ,label="v")
plt.grid()
plt.legend()
# plt.xlabel("meshsize in multiples of $h_0$")
plt.ylabel("$L_2$-error")
# plt.xscale("log")
plt.yscale("log")
# plt.savefig("../out/dirichlet_conv_TGV_Re500_timedep.pdf")
```
