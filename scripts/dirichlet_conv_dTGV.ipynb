```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

- dynamic TGV, order 2, dt=0.0001, Re1
- dynamic TGV, order 2, dt=0.0001, invisc

# dynamic TGV, dt=0.01

```python
filename = "dirichlet-conv-Re1-dTGV"

data   = np.genfromtxt("../out/rawdata/"+filename+".txt", delimiter=',').flatten()
refstep = data[0::4][-5:]
u =       data[1::4][-5:]
v =       data[2::4][-5:]
uvdiff =  data[3::4][-5:]

plt.plot(refstep,u,"m.-",label= "u")
plt.plot(refstep,v,"co-",label="v")
plt.plot(refstep,uvdiff,"b.:",label="u-v")
plt.plot(refstep,[0.25*refstep[i]**1 for i in range(len(refstep))],"k-.",label="$O(h^1)$")
plt.plot(refstep,[0.05*refstep[i]**1 for i in range(len(refstep))],"k-.")
plt.plot(refstep,[0.01*refstep[i]**2 for i in range(len(refstep))],"k--",label="$O(h^2)$")
plt.legend()
plt.yscale("log")
plt.xscale("log")
plt.ylabel("$L^2$-error")
plt.xlabel("$h$")
plt.tight_layout()  
plt.grid()
plt.savefig("../out/plots/"+filename+".pdf")
```
