```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

### dirichlet: ess BC on dual space, Re=$\infty$

```python
data = genfromtxt("../out/dirichlet_dual_conv_inviscid.txt", delimiter=' ').flatten()
v =        data[1::5]
u =        data[2::5]
uvdiff =   data[3::5]
time =     data[4::5]
meshsize = data[0::5]
```

```python
plt.plot(meshsize,uvdiff,"c.-",label="$u-v$")
plt.plot(meshsize,u,"m.-"     ,label="$u-u_{exact}$")
plt.plot(meshsize,v,"k.--"    ,label="$v-v_{exact}$")
plt.grid()
plt.legend()
plt.xlabel("mesh refinement steps")
plt.ylabel("$L_2$-error")
plt.xticks(np.arange(0, 4, step=1))
# plt.xscale("log")
plt.yscale("log")
plt.savefig("../out/plots/dirichlet_dual_conv_inviscid.pdf")
```
