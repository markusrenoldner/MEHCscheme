```python
import numpy as np
import matplotlib.pyplot as plt
```

### convergence, TGV wikipedia, Re=500 and time-dependent, dt=0.05, tmax=0.15

```python
data=[
0.374856
,1.21622
,0.902688
,554.365
,1
,0.160748
,0.607303
,0.47397
,17674
,0.5
]

v =      data[0::5]
u =      data[1::5]
uvdiff = data[2::5]
time =   data[3::5]
meshsize = data[4::5]
```

```python
plt.plot(meshsize,uvdiff,"c.-",label="u-v")
plt.plot(meshsize,u,"m.-"     ,label="u")
plt.plot(meshsize,v,"k.--"    ,label="v")
plt.grid()
plt.legend()
plt.xlabel("meshsize in multiples of $h_0$")
plt.ylabel("$L_2$-error")
plt.xscale("log")
plt.yscale("log")
plt.savefig("../out/conv_TGV_Re500_timedep.pdf")
```
