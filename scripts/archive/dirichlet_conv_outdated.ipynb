```python
data = np.genfromtxt("../out/rawdata/dirichlet-primal-test.txt", delimiter=',').flatten()
dataD = np.genfromtxt("../out/rawdata/dirichlet-dual-test.txt", delimiter=',').flatten()


refstep = data[0::2][-5:]
u =       data[1::2][-5:]
v =       dataD[1::2][-5:]

plt.plot(refstep,u,"m.-",label="u")
plt.plot(refstep,v,"c.-",label="v")
plt.plot(refstep,[0.05*refstep[i]**1 for i in range(len(refstep))],"k--",label="h^1")
plt.plot(refstep,[0.02*refstep[i]**2 for i in range(len(refstep))],"k-.",label="h^2")
plt.plot(refstep,[0.01*refstep[i]**3 for i in range(len(refstep))],"k-.",label="h^3")

plt.legend()
plt.title("TEST")
plt.yscale("log")
plt.xscale("log")
plt.ylabel("L2error")
plt.xlabel("h")
plt.tight_layout()
plt.grid()
```

# decoupled dual system

```python
data100 = np.genfromtxt("../out/rawdata/dirichlet-dual-conv-Re100.txt", delimiter=',').flatten()
data1   = np.genfromtxt("../out/rawdata/dirichlet-dual-conv-Re1.txt", delimiter=',').flatten()
datainv = np.genfromtxt("../out/rawdata/dirichlet-dual-conv-invisc.txt", delimiter=',').flatten()


refstep =   data1[0::2][-5:]
u100 =    data100[1::2][-5:]
u1 =        data1[1::2][-5:]
uinv =    datainv[1::2][-5:]

plt.plot(refstep,u100,"m.-",label="u, Re=100")
plt.plot(refstep,u1,"c.--",label="u, Re=1")
plt.plot(refstep,uinv,"k.--",label="u, inviscid")

plt.legend()
plt.title("Dirichlet, dual")
plt.yscale("log")
plt.xscale("log")
plt.ylabel("L2error")
plt.xlabel("h")
plt.tight_layout()
plt.grid()
```
