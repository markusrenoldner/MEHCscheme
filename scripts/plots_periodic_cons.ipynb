```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import genfromtxt
```

### conservation: periodic domain, Re=$\infty$, $\Delta t=0.05$

```python
data = genfromtxt("../out/cons_invisc.txt", delimiter=',').flatten()

time = data[0::8]
m1 = data[1::8]
m2 = data[2::8]
k1 = data[3::8]
k2 = data[4::8]
h1 = data[5::8]
h2 = data[6::8]
```

```python
plt.plot(time,m1,label="div$(u)$")
plt.plot(time,m2,label="div$(v)$")
plt.legend()
plt.grid()
plt.ylim(0,0.000000000001)
plt.xlabel("time")
plt.ylabel("div$(u)$, div$(v)$")
plt.savefig("../out/cons_invisc_div.pdf")
```

```python
plt.plot(time,k1,label="$K_1$")
plt.plot(time,k2,label="$K_2$")
plt.legend()
plt.grid()
plt.ylim(-0.00000000001,0.00000000001)
plt.xlabel("time")
plt.ylabel("$K_1$, $K_2$")
plt.savefig("../out/cons_invisc_K.pdf")
```

```python
plt.plot(time,h1,label="$H_1$")
plt.plot(time,h2,label="$H_2$")
plt.legend()
plt.grid()
plt.ylim(-0.00000000001,0.00000000001)
plt.xlabel("time")
plt.ylabel("$H_1$, $H_2$")
plt.savefig("../out/cons_invisc_H.pdf")
```

### conservation: periodic domain, Re=$100$, $\Delta t=0.05$

```python
data = genfromtxt("../out/cons_visc.txt", delimiter=',').flatten()

time = data[0::6]
k1 = data[1::6]
k2 = data[2::6]
h1 = data[3::6]
h2 = data[4::6]
```

```python
plt.plot(time,k1,label="$\Delta K_1-2E_2/$Re")
plt.plot(time,k2,label="$\Delta K_2-2E_1/$Re")
plt.legend()
plt.grid()
plt.ylim(-0.000000000001,0.000000000001)
# plt.ylim(-2,2)
plt.xlabel("time")
plt.ylabel("$K_1$, $K_2$")
plt.savefig("../out/cons_visc_K.pdf")
```
