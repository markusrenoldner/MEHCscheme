# manuf sol from the paper sec 5.1.1

```python

 
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
z = np.arange(-1, 1, 0.1)
X,  Z = np.meshgrid(x,  z)

pi = np.pi

ux,uy,uz = np.array([np.cos(2*Z*pi),np.sin(2*pi*Z),np.sin(2*pi*X)])
 
# color = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
M = np.sqrt(ux*ux+uz*uz) # magnitude
plt.quiver(X, Z, ux,uz,M,cmap=plt.cm.jet)
plt.xticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.yticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.colorbar(cmap=plt.cm.jet)
plt.tight_layout()
# plt.show()
plt.savefig("../out/plots/periodic-cons-u0-xz.pdf")
```

# wikipedia TGV solution with Re=\infty 

=> in this case u is time-independent and curl(f) = 0

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.arange(0, 2*np.pi, 0.2)
y = np.arange(0, 2*np.pi, 0.2)
X, Y = np.meshgrid(x, y)

ux,uy = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
 
# color = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
M = np.sqrt(ux*ux+uy*uy) # magnitude
plt.quiver(X, Y, ux,uy,M,cmap=plt.cm.jet)
plt.xticks(np.arange(0, 2*np.pi+0.01, np.pi), ['0','π','2π'])
plt.yticks(np.arange(0, 2*np.pi+0.01, np.pi), ['0','π','2π'])
plt.colorbar(cmap=plt.cm.jet)
plt.tight_layout()
plt.show()
```

# rescaling for [-1,1]^3

```python
x = np.arange(-1,1, 0.2/np.pi)
y = np.arange(-1,1, 0.2/np.pi)
X, Y = np.meshgrid(x, y)

ux,uy = np.array([np.cos(X*np.pi)*np.sin(Y*np.pi),
                 -np.sin(X*np.pi)*np.cos(Y*np.pi)])
 
# color = np.array([np.cos(X)*np.sin(Y),-np.sin(X)*np.cos(Y)])
M = np.sqrt(ux*ux+uy*uy) # magnitude
plt.quiver(X, Y, ux,uy,M,cmap=plt.cm.jet)
plt.xticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.yticks(np.arange(-1, 1+0.01, 1), ['-1','0','1'])
plt.colorbar(cmap=plt.cm.jet)
# plt.show()
plt.tight_layout()
plt.savefig("../out/plots/periodic-conv-u0-TGV.pdf")
```

# rescaling for non-periodic reference cube [0,1]^3

```python
x = np.arange(0,1, 0.2/np.pi)
y = np.arange(0,1, 0.2/np.pi)
X, Y = np.meshgrid(x, y)

ux,uy = np.array([np.cos(X*np.pi)*np.sin(Y*np.pi),
                 -np.sin(X*np.pi)*np.cos(Y*np.pi)])

M = np.sqrt(ux*ux+uy*uy) # magnitude
plt.title("velocity")
plt.quiver(X, Y, ux,uy,M,cmap=plt.cm.jet)
plt.xticks(np.arange(0, 1+0.01, 1), ['0','1'])
plt.yticks(np.arange(0, 1+0.01, 1), ['0','1'])
plt.colorbar(cmap=plt.cm.jet)
plt.tight_layout()
plt.show()
```

# magnitude of vorticity of TGV

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,1, 0.05/np.pi)
y = np.arange(0,1, 0.05/np.pi)
X, Y = np.meshgrid(x, y)
ux,uy = np.array([np.cos(X*np.pi)*np.sin(Y*np.pi),
                 -np.sin(X*np.pi)*np.cos(Y*np.pi)])

curl = -2*np.pi*np.cos(np.pi*X)*np.cos(np.pi*Y)

plt.title("vorticity")
plt.pcolormesh(X,Y, curl)
plt.colorbar()
plt.xticks(np.arange(0, 1+0.01, 1), ['0','1'])
plt.yticks(np.arange(0, 1+0.01, 1), ['0','1'])
# plt.colorbar(cmap=plt.cm.jet)
plt.tight_layout()
plt.show()
```
