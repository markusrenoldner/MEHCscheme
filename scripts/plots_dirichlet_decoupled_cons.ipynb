### dirichlet: ess BC on dual space, Re=$\infty$

```python
data = genfromtxt("../out/dirichlet_dual_cons_inviscid.txt", delimiter=',').flatten()

time = data[0::8]
m1 = data[1::8]
m2 = data[2::8]
k1 = data[3::8]
k2 = data[4::8]
h1 = data[5::8]
h2 = data[6::8]
```

```python
plt.plot(time,m1,label="div$(u)$")
plt.plot(time,m2,label="div$(v)$")
plt.legend()
plt.grid()
# plt.ylim(0,0.000000000001)
# plt.yscale("log")
plt.xlabel("time")
plt.ylabel("div$(u)$, div$(v)$")
# plt.savefig("../out/plots/ .... .pdf")
```

```python
plt.plot(time,k1,label="$K_1$")
plt.plot(time,k2,label="$K_2$")
plt.legend()
plt.grid()
# plt.ylim(0,0.000000000001)
# plt.yscale("log")
plt.xlabel("time")
plt.ylabel("$K_1$, $K_2$")
# plt.savefig("../out/plots/ .... .pdf")
```

```python
plt.plot(time,h1,label="H$(u)$")
plt.plot(time,h2,label="H$(v)$")
plt.legend()
plt.grid()
# plt.ylim(0,0.000000000001)
# plt.yscale("log")
plt.xlabel("time")
plt.ylabel("$H_1$, $H_2$")
# plt.savefig("../out/plots/ .... .pdf")
```
